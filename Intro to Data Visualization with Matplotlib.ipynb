# plotting two time series together

```python
fig, ax = plt.subplots()

ax.plot(climate_change.index, climate_change['co2'])
ax.plot(climate_change.index, climate_change['relative_temp'])
ax.set_xlabel('Time')
ax.set_ylabel('C02 (ppm) / Relative Temperature')
plt.show()

#this does not give an accurate representation of the plots as they differ in the y axis
```

```python
# using twin axes [ax.twinx() ]
#this is used when we want to make comparison of two sets of data with the same x-axis but different y-axis
```

```python
#adding colors to easily differentiate the plots

fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('C02 (ppm)', color='b')
ax.tick_params('y', colors='b')

ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change['relative_temp'], color='r')
ax2.set_ylabel('Relative Temperature (celsius)', color='r')
ax2.tick_params('y', colors='r')
plt.show()
```

```python
#using a simple function for plotting two time series
fig, ax = plt.subplots()
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y', color=color)
```

```python
fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change['co2'], 'blue', 'Time', 'C02 (ppm)')

ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 'red', 'Time', 'Relative Temperature (celsius)')
plt.show()
```

# annotating/pointing time series data

this is used to draw attention to a part of a plot
for example drawing an arrow to a part of of the plot and including a text to explain it

```python
#annotating/pointing the aspect of the section of the plot where temperature exceeded 1 degrees

fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change['co2'], 'blue', 'Time', 'C02 (ppm)')

ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 'red', 'Time', 'Relative Temperature (celsius)')
ax2.annotate('>1 degree', xy=[pd.Timestamp('2015-10-06'), 1], 
             xytext = [pd.Timestamp('2008-10-6'), -0.2], arrowprops={})
plt.show()

#xy function is used to select the year in when temperature exceeded 1 degrees. x is for timestamp and y is for temperature
#xytext is used to select where we want to position where we want the annotation to be displayed to avoid not being seen clearly
#arrowprops is used to shpw the arrow
```

```python
#customizing arrow properties using the arrowprops={} function

fig, ax = plt.subplots()
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 'red', 'Time', 'Relative Temperature (celsius)')
ax2.annotate('>1 degree', xy=[pd.Timestamp('2015-10-06'), 1], 
             xytext = [pd.Timestamp('2008-10-6'), -0.2], arrowprops={'arrowstyle':'->', 'color':'gray'})
plt.show()
```

```python
#stacking/visualizing data on top each other

fig, ax = plt.subplots()

ax.bar(medals.index, medals['Gold'])
ax.bar(medals.index, medals['Silver'], bottom=medals['Gold'])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel('Number of Medals')
plt.legend(['Gold', 'Silver']) #alternative for adding color description
plt.show()
```

```python
#label is used to add color description as to which color represents the medals 
#add ax.legend() afterwards for the function to take action
fig, ax = plt.subplots()

ax.bar(medals.index, medals['Gold'], label='Gold')
ax.bar(medals.index, medals['Silver'], bottom = medals['Gold'], label='Silver')
ax.bar(medals.index, medals['Bronze'], bottom = medals['Gold'] + medals['Silver'], label='Bronze')
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel('Number of Medals')
ax.legend()
plt.show()
```

```python
#alternative for adding color description

fig, ax = plt.subplots()

ax.bar(medals.index, medals['Gold'])
ax.bar(medals.index, medals['Silver'], bottom = medals['Gold'])
ax.bar(medals.index, medals['Bronze'], bottom = medals['Gold'] + medals['Silver'])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel('Number of Medals')
ax.legend(['Gold', 'Silver', 'Bronze'])
plt.show()
```

# quantitative comparisons: histogram

```python
medalist = pd.read_csv('summer2016.csv', index_col=0)
medalist.head()
```

```python
mens_rowing = medalist[(medalist.Sex=='M') & (medalist.Sport=='Rowing')]
mens_rowing.head()
```

```python
mens_gymnastics = medalist[(medalist.Sex=='M') & (medalist.Sport=='Gymnastics')]
mens_gymnastics.head()
```

```python
#comparing the average height between mens rowing and men gymnastics
fig, ax=plt.subplots()

ax.bar('Rowing', mens_rowing['Height'].mean())
ax.bar('Gymnastics', mens_gymnastics['Height'].mean())
ax.set_ylabel('Height (cm)')
plt.show()
```

```python
#histogram using axes subplots

fig, ax=plt.subplots()

ax.hist(mens_rowing['Height'])
ax.hist(mens_gymnastics['Height'])
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of Observations')
plt.legend(['Mens Rowling', 'Mens Gymnastics'])
plt.show()

#the x axis have been converted to represent the heights and the y axis represents the frequency ()
#the number of bars or bins in automatically set at 10
```

```python
#alternative for drawing histogram
mens_rowing['Height'].hist()
mens_gymnastics['Height'].hist(alpha=0.5) #alpha is used to reduce the opacity of the bar
plt.xlabel('Height (cm)')
plt.ylabel('# of Observations')
plt.legend(['Mens Rowling', 'Mens Gymnastics'])
plt.show()
```

```python
#changing/customizing the number of bars/bins to five(5)

fig, ax=plt.subplots()

ax.hist(mens_rowing['Height'], bins=5)
ax.hist(mens_gymnastics['Height'], bins=5, histtype='step') #histtype is used to change the transparency/opacity of the bar
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of Observations')
plt.legend(['Mens Rowling', 'Mens Gymnastics'])
plt.show()
```

```python
#changing/customizing the number of bars/bins

fig, ax=plt.subplots()

ax.hist(mens_rowing['Height'], bins=[150, 160, 170, 180, 190, 200, 210])
ax.hist(mens_gymnastics['Height'], bins=[150, 160, 170, 180, 190, 200, 210], histtype='step')
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of Observations')
plt.legend(['Mens Rowling', 'Mens Gymnastics'])
plt.show()
```

# statistical plotting 

Statistical plotting is a set of methods for using visualizations to make comparisons. There are two(2) techniques:

1. using Error Bars 
Error bars are graphical representations of the variability of data and used on graphs to indicate the error or uncertainty in a reported measurement. They give a general idea of how precise a measurement is, or conversely, how far from the reported value the true (error free) value might be.

NB: a short error bar shows reliablity or less deviation from the mean and vice versa

2. using Box PLots
n descriptive statistics, a box plot or boxplot is a method for graphically depicting groups of numerical data through their quartiles. Box plots may also have lines extending from the boxes indicating variability outside the upper and lower quartiles, hence the terms box-and-whisker plot and box-and-whisker diagram.

```python
# Error Bars in a bar chart

fig, ax=plt.subplots()

ax.bar('Rowing', mens_rowing['Height'].mean(), yerr=mens_rowing['Height'].std())
ax.bar('Gymnastics', mens_gymnastics['Height'].mean(), yerr=mens_gymnastics['Height'].std())
ax.set_ylabel('Height (cm)')
plt.show()
```

```python
# Error Bars in a plot

fig, ax=plt.subplots()

ax.errorbar(seattle_weather['DATE'], seattle_weather['MLY-TAVG-NORMAL'], yerr=seattle_weather['MLY-TAVG-STDDEV'])
ax.errorbar(austin_weather['DATE'], austin_weather['MLY-TAVG-NORMAL'], yerr=austin_weather['MLY-TAVG-STDDEV'])
ax.set_ylabel('Temperature (Fahrenheit)')
```

```python
#using Box Plots

fig, ax=plt.subplots()

ax.boxplot([mens_rowing['Height'], mens_gymnastics['Height']])
ax.set_xticklabels(['Rowing', 'Gymnastics'])
ax.set_ylabel('Height (cm)')
```

```python
#getting unique values of a column
#when we have a long data frame where we cannot tell how many unique values we have in a specific column, we use the unique function to check it out

sports = medalist['Sport'].unique()
sports

#the result shows use the branches of sports we have in the dataframe
```
